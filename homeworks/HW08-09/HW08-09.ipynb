```python
# ============================================================================
# HW08-09: PyTorch MLP - Регуляризация и Оптимизация
# ============================================================================
# Этот ноутбук выполняет все требования домашней работы:
# - Часть A (S08): Регуляризация (Dropout, BatchNorm, EarlyStopping)
# - Часть B (S09): Оптимизация (Learning Rate, Adam vs SGD, Weight Decay)
# ============================================================================

# ============================================================================
# 1. ИМПОРТЫ И ОБЩИЕ НАСТРОЙКИ
# ============================================================================

import os
import json
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, Subset
from torchvision import datasets, transforms

# Проверка версий
print(f"torch: {torch.__version__}")
print(f"torchvision: {datasets.__version__ if hasattr(datasets, '__version__') else 'N/A'}")

# ============================================================================
# 2. ВОСПРОИЗВОДИМОСТЬ И УСТРОЙСТВО
# ============================================================================

SEED = 42

def set_seed(seed: int = 42) -> None:
    """Фиксация seed для воспроизводимости результатов."""
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

set_seed(SEED)

# Выбор устройства (GPU/CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {device}")

# ============================================================================
# 3. НАСТРОЙКИ ЭКСПЕРИМЕНТА
# ============================================================================

# Быстрый режим для ускорения (можно поставить False для полного обучения)
FAST_MODE = True

# Параметры данных
DATA_CONFIG = {
    'dataset_name': 'EMNIST',  # Вместо 'KMNIST'
    'batch_size': 128 if device.type == 'cuda' else 64,
    'val_fraction': 0.15,
    'num_workers': 0,
    'fast_mode': True,
    'max_train_samples': 12000,
    'max_val_samples': 2000,
    'max_test_samples': 3000,
}

# Параметры модели
MODEL_CONFIG = {
    'hidden_dims': (256, 128),
    'activation': 'ReLU',
    'dropout_p': 0.3,
    'use_batchnorm': True,
}

# Параметры обучения
TRAIN_CONFIG = {
    'epochs': 30,
    'lr': 1e-3,
    'weight_decay': 0,
    'momentum': 0.9,
    'early_stopping_patience': 5,
    'early_stopping_min_delta': 0.0005,
}

# ============================================================================
# 4. ЗАГРУЗКА ДАННЫХ
# ============================================================================

def load_dataset(config: dict, seed: int = 42):
    """Загружает датасет и создаёт DataLoader для train/val/test."""

    # Трансформации
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,)),  # Нормализация для KMNIST
    ])

    # Загрузка датасета
    if config['dataset_name'] == 'KMNIST':
        train_full = datasets.KMNIST(root='./data', train=True, download=True, transform=transform)
        test_ds = datasets.KMNIST(root='./data', train=False, download=True, transform=transform)
        input_dim = 28 * 28
        num_classes = 10
    elif config['dataset_name'] == 'EMNIST':
        train_full = datasets.EMNIST(root='./data', split='balanced', train=True, download=True, transform=transform)
        test_ds = datasets.EMNIST(root='./data', split='balanced', train=False, download=True, transform=transform)
        input_dim = 28 * 28
        num_classes = 47
    elif config['dataset_name'] == 'CIFAR10':
        transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
        ])
        train_full = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
        test_ds = datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)
        input_dim = 32 * 32 * 3
        num_classes = 10
    else:
        raise ValueError(f"Unknown dataset: {config['dataset_name']}")

    # Ограничение размера для FAST_MODE
    if config['max_train_samples']:
        train_full = Subset(train_full, range(min(len(train_full), config['max_train_samples'])))
    if config['max_test_samples']:
        test_ds = Subset(test_ds, range(min(len(test_ds), config['max_test_samples'])))

    # Разбиение train/val
    val_size = int(len(train_full) * config['val_fraction'])
    train_size = len(train_full) - val_size

    generator = torch.Generator().manual_seed(seed)
    train_ds, val_ds = random_split(train_full, [train_size, val_size], generator=generator)

    # Создание DataLoader
    train_loader = DataLoader(
        train_ds, batch_size=config['batch_size'], shuffle=True,
        num_workers=config['num_workers'], pin_memory=(device.type == 'cuda')
    )
    val_loader = DataLoader(
        val_ds, batch_size=config['batch_size'], shuffle=False,
        num_workers=config['num_workers'], pin_memory=(device.type == 'cuda')
    )
    test_loader = DataLoader(
        test_ds, batch_size=config['batch_size'], shuffle=False,
        num_workers=config['num_workers'], pin_memory=(device.type == 'cuda')
    )

    return train_loader, val_loader, test_loader, input_dim, num_classes

# Загрузка данных
train_loader, val_loader, test_loader, INPUT_DIM, NUM_CLASSES = load_dataset(DATA_CONFIG, SEED)

print(f"Dataset: {DATA_CONFIG['dataset_name']}")
print(f"Input dim: {INPUT_DIM}, Num classes: {NUM_CLASSES}")
print(f"Train batches: {len(train_loader)}, Val batches: {len(val_loader)}, Test batches: {len(test_loader)}")

# ============================================================================
# 5. SANITY CHECK ДАННЫХ
# ============================================================================

x_batch, y_batch = next(iter(train_loader))
print(f"\nSanity Check:")
print(f"x_batch shape: {x_batch.shape}, dtype: {x_batch.dtype}")
print(f"y_batch shape: {y_batch.shape}, dtype: {y_batch.dtype}")
print(f"x_batch range: [{x_batch.min():.3f}, {x_batch.max():.3f}]")
print(f"y_batch range: [{y_batch.min()}, {y_batch.max()}]")

# ============================================================================
# 6. МОДЕЛЬ MLP
# ============================================================================

class MLP(nn.Module):
    """Многослойный перцептрон с поддержкой Dropout и BatchNorm."""

    def __init__(
        self,
        input_dim: int,
        num_classes: int,
        hidden_dims: tuple = (256, 128),
        dropout_p: float = 0.0,
        use_batchnorm: bool = False,
        activation: str = 'ReLU'
    ):
        super().__init__()

        layers = []
        prev_dim = input_dim

        # Активация
        if activation == 'ReLU':
            act_fn = nn.ReLU
        elif activation == 'Tanh':
            act_fn = nn.Tanh
        elif activation == 'Sigmoid':
            act_fn = nn.Sigmoid
        else:
            act_fn = nn.ReLU

        # Скрытые слои
        for h in hidden_dims:
            layers.append(nn.Linear(prev_dim, h))
            if use_batchnorm:
                layers.append(nn.BatchNorm1d(h))
            layers.append(act_fn())
            if dropout_p > 0:
                layers.append(nn.Dropout(p=dropout_p))
            prev_dim = h

        # Выходной слой
        layers.append(nn.Linear(prev_dim, num_classes))

        self.net = nn.Sequential(*layers)

    def forward(self, x):
        # Flatten для изображений
        if x.dim() > 2:
            x = x.view(x.size(0), -1)
        return self.net(x)

# Тест модели
test_model = MLP(INPUT_DIM, NUM_CLASSES, **MODEL_CONFIG).to(device)
with torch.no_grad():
    out = test_model(x_batch.to(device))
print(f"\nModel output shape: {out.shape}")
print(f"Model parameters: {sum(p.numel() for p in test_model.parameters()):,}")

# ============================================================================
# 7. ФУНКЦИИ ОБУЧЕНИЯ И ОЦЕНКИ
# ============================================================================

def train_one_epoch(model, loader, optimizer, criterion, device):
    """Одна эпоха обучения."""
    model.train()
    total_loss = 0.0
    total_correct = 0
    total_samples = 0

    for x, y in loader:
        x = x.to(device, non_blocking=True)
        y = y.to(device, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)
        logits = model(x)
        loss = criterion(logits, y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * x.size(0)
        total_correct += (logits.argmax(dim=1) == y).sum().item()
        total_samples += x.size(0)

    return total_loss / total_samples, total_correct / total_samples

def evaluate(model, loader, criterion, device):
    """Оценка на валидации/тесте."""
    model.eval()
    total_loss = 0.0
    total_correct = 0
    total_samples = 0

    with torch.no_grad():
        for x, y in loader:
            x = x.to(device, non_blocking=True)
            y = y.to(device, non_blocking=True)

            logits = model(x)
            loss = criterion(logits, y)

            total_loss += loss.item() * x.size(0)
            total_correct += (logits.argmax(dim=1) == y).sum().item()
            total_samples += x.size(0)

    return total_loss / total_samples, total_correct / total_samples

class EarlyStopping:
    """Ранняя остановка обучения."""

    def __init__(self, patience: int = 5, min_delta: float = 0.0005):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.best_state = None

    def __call__(self, model, val_loss):
        if self.best_loss is None or val_loss < self.best_loss - self.min_delta:
            self.best_loss = val_loss
            self.best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
            self.counter = 0
            return False
        else:
            self.counter += 1
            return self.counter >= self.patience

    def restore_best(self, model):
        if self.best_state is not None:
            model.load_state_dict(self.best_state)

def fit(model, train_loader, val_loader, optimizer, criterion, device,
        epochs=20, early_stopping=None, verbose=True):
    """Полный цикл обучения."""
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

    for epoch in range(epochs):
        train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, criterion, device)
        val_loss, val_acc = evaluate(model, val_loader, criterion, device)

        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)

        if verbose:
            print(f"Epoch {epoch+1}/{epochs} | "
                  f"train loss={train_loss:.4f}, acc={train_acc:.4f} | "
                  f"val loss={val_loss:.4f}, acc={val_acc:.4f}")

        if early_stopping and early_stopping(model, val_loss):
            if verbose:
                print(f"Early stopping at epoch {epoch+1}")
            early_stopping.restore_best(model)
            break

    return history

# ============================================================================
# 8. УТИЛИТЫ ДЛЯ СОХРАНЕНИЯ РЕЗУЛЬТАТОВ
# ============================================================================

# Создание папок для артефактов
ARTIFACTS_DIR = Path('artifacts')
FIGURES_DIR = ARTIFACTS_DIR / 'figures'
ARTIFACTS_DIR.mkdir(exist_ok=True)
FIGURES_DIR.mkdir(exist_ok=True)

# Список для хранения результатов всех экспериментов
all_runs = []

def save_experiment_result(exp_id, config, history, model=None, is_best=False):
    """Сохраняет результаты эксперимента в runs.csv."""
    best_val_acc = max(history['val_acc']) if history['val_acc'] else 0
    best_val_loss = min(history['val_loss']) if history['val_loss'] else 0
    epochs_trained = len(history['val_acc'])

    result = {
        'experiment_id': exp_id,
        'dataset': DATA_CONFIG['dataset_name'],
        'seed': SEED,
        'model_summary': f"hidden={config.get('hidden_dims', MODEL_CONFIG['hidden_dims'])}/"
                        f"dropout={config.get('dropout_p', 0)}/"
                        f"batchnorm={config.get('use_batchnorm', False)}",
        'optimizer': config.get('optimizer', 'Adam'),
        'lr': config.get('lr', TRAIN_CONFIG['lr']),
        'momentum': config.get('momentum', 0),
        'weight_decay': config.get('weight_decay', 0),
        'epochs_trained': epochs_trained,
        'best_val_accuracy': best_val_acc,
        'best_val_loss': best_val_loss,
    }
    all_runs.append(result)

    # Сохранение лучшей модели
    if is_best and model is not None:
        torch.save(model.state_dict(), ARTIFACTS_DIR / 'best_model.pt')

        # Сохранение конфига
        best_config = {
            'dataset': DATA_CONFIG['dataset_name'],
            'seed': SEED,
            'model': {
                'input_dim': INPUT_DIM,
                'num_classes': NUM_CLASSES,
                'hidden_dims': config.get('hidden_dims', MODEL_CONFIG['hidden_dims']),
                'dropout_p': config.get('dropout_p', 0),
                'use_batchnorm': config.get('use_batchnorm', False),
                'activation': config.get('activation', 'ReLU'),
            },
            'training': {
                'optimizer': config.get('optimizer', 'Adam'),
                'lr': config.get('lr', TRAIN_CONFIG['lr']),
                'weight_decay': config.get('weight_decay', 0),
                'batch_size': DATA_CONFIG['batch_size'],
                'epochs': epochs_trained,
            }
        }
        with open(ARTIFACTS_DIR / 'best_config.json', 'w') as f:
            json.dump(best_config, f, indent=2)

    return result

def plot_history(history, title='', filename=None):
    """Построение графиков обучения."""
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    # Loss
    axes[0].plot(history['train_loss'], label='Train Loss')
    axes[0].plot(history['val_loss'], label='Val Loss')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].set_title(f'Loss - {title}')
    axes[0].legend()
    axes[0].grid(True)

    # Accuracy
    axes[1].plot(history['train_acc'], label='Train Acc')
    axes[1].plot(history['val_acc'], label='Val Acc')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy')
    axes[1].set_title(f'Accuracy - {title}')
    axes[1].legend()
    axes[1].grid(True)

    plt.tight_layout()
    if filename:
        plt.savefig(filename, dpi=150, bbox_inches='tight')
        print(f"Saved: {filename}")
    plt.show()

# ============================================================================
# 9. ЧАСТЬ A (S08): РЕГУЛЯРИЗАЦИЯ (E1-E4)
# ============================================================================

print("\n" + "="*60)
print("ЧАСТЬ A: РЕГУЛЯРИЗАЦИЯ (E1-E4)")
print("="*60)

criterion = nn.CrossEntropyLoss()
best_exp_id = None
best_val_acc = 0
best_history = None
best_model_config = None

# ----------------------------------------------------------------------------
# E1: Baseline (без регуляризации)
# ----------------------------------------------------------------------------
print("\n--- E1: Baseline (без Dropout/BatchNorm) ---")
set_seed(SEED)

# Параметры ТОЛЬКО для модели
model_config_e1 = {
    'hidden_dims': (256, 128),
    'dropout_p': 0.0,
    'use_batchnorm': False,
    'activation': 'ReLU',
}

# Параметры для обучения
train_config_e1 = {
    'optimizer': 'Adam',
    'lr': 1e-3,
    'weight_decay': 0,
}

model_e1 = MLP(INPUT_DIM, NUM_CLASSES, **model_config_e1).to(device)

# Создаём оптимизатор отдельно
if train_config_e1['optimizer'] == 'Adam':
    optimizer_e1 = optim.Adam(model_e1.parameters(),
                               lr=train_config_e1['lr'],
                               weight_decay=train_config_e1['weight_decay'])
elif train_config_e1['optimizer'] == 'SGD':
    optimizer_e1 = optim.SGD(model_e1.parameters(),
                              lr=train_config_e1['lr'],
                              momentum=0.9,
                              weight_decay=train_config_e1['weight_decay'])

history_e1 = fit(model_e1, train_loader, val_loader, optimizer_e1, criterion, device,
                 epochs=TRAIN_CONFIG['epochs'], verbose=True)

# Для сохранения в runs.csv объединяем конфиги
config_e1_full = {**model_config_e1, **train_config_e1}
save_experiment_result('E1', config_e1_full, history_e1)
plot_history(history_e1, 'E1 Baseline', FIGURES_DIR / 'curves_e1.png')

if history_e1['val_acc'][-1] > best_val_acc:
    best_val_acc = history_e1['val_acc'][-1]
    best_exp_id = 'E1'
    best_history = history_e1
    best_model_config = model_config_e1.copy()
    best_train_config = train_config_e1.copy()

# ----------------------------------------------------------------------------
# E2: Dropout
# ----------------------------------------------------------------------------
print("\n--- E2: Dropout (p=0.3) ---")
set_seed(SEED)

model_config_e2 = {
    'hidden_dims': (256, 128),
    'dropout_p': 0.3,
    'use_batchnorm': False,
    'activation': 'ReLU',
}

train_config_e2 = {
    'optimizer': 'Adam',
    'lr': 1e-3,
    'weight_decay': 0,
}

model_e2 = MLP(INPUT_DIM, NUM_CLASSES, **model_config_e2).to(device)
optimizer_e2 = optim.Adam(model_e2.parameters(), lr=train_config_e2['lr'])
history_e2 = fit(model_e2, train_loader, val_loader, optimizer_e2, criterion, device,
                 epochs=TRAIN_CONFIG['epochs'], verbose=True)

config_e2_full = {**model_config_e2, **train_config_e2}
save_experiment_result('E2', config_e2_full, history_e2)
plot_history(history_e2, 'E2 Dropout', FIGURES_DIR / 'curves_e2.png')

if history_e2['val_acc'][-1] > best_val_acc:
    best_val_acc = history_e2['val_acc'][-1]
    best_exp_id = 'E2'
    best_history = history_e2
    best_model_config = model_config_e2.copy()
    best_train_config = train_config_e2.copy()

# ----------------------------------------------------------------------------
# E3: BatchNorm
# ----------------------------------------------------------------------------
print("\n--- E3: BatchNorm ---")
set_seed(SEED)

model_config_e3 = {
    'hidden_dims': (256, 128),
    'dropout_p': 0.0,
    'use_batchnorm': True,
    'activation': 'ReLU',
}

train_config_e3 = {
    'optimizer': 'Adam',
    'lr': 1e-3,
    'weight_decay': 0,
}

model_e3 = MLP(INPUT_DIM, NUM_CLASSES, **model_config_e3).to(device)
optimizer_e3 = optim.Adam(model_e3.parameters(), lr=train_config_e3['lr'])
history_e3 = fit(model_e3, train_loader, val_loader, optimizer_e3, criterion, device,
                 epochs=TRAIN_CONFIG['epochs'], verbose=True)

config_e3_full = {**model_config_e3, **train_config_e3}
save_experiment_result('E3', config_e3_full, history_e3)
plot_history(history_e3, 'E3 BatchNorm', FIGURES_DIR / 'curves_e3.png')

if history_e3['val_acc'][-1] > best_val_acc:
    best_val_acc = history_e3['val_acc'][-1]
    best_exp_id = 'E3'
    best_history = history_e3
    best_model_config = model_config_e3.copy()
    best_train_config = train_config_e3.copy()

# ----------------------------------------------------------------------------
# E4: EarlyStopping (лучший из E2/E3 + ES)
# ----------------------------------------------------------------------------
print("\n--- E4: EarlyStopping (лучший из E2/E3) ---")

# Выбираем лучший из E2/E3
if history_e2['val_acc'][-1] >= history_e3['val_acc'][-1]:
    print("Выбираем конфигурацию E2 (Dropout) для E4")
    model_config_e4 = model_config_e2.copy()
else:
    print("Выбираем конфигурацию E3 (BatchNorm) для E4")
    model_config_e4 = model_config_e3.copy()

set_seed(SEED)
train_config_e4 = {
    'optimizer': 'Adam',
    'lr': 1e-3,
    'weight_decay': 0,
}

model_e4 = MLP(INPUT_DIM, NUM_CLASSES, **model_config_e4).to(device)
optimizer_e4 = optim.Adam(model_e4.parameters(), lr=train_config_e4['lr'])
early_stopping = EarlyStopping(
    patience=TRAIN_CONFIG['early_stopping_patience'],
    min_delta=TRAIN_CONFIG['early_stopping_min_delta']
)

history_e4 = fit(
    model_e4, train_loader, val_loader, optimizer_e4, criterion, device,
    epochs=50,  # Больше эпох для early stopping
    early_stopping=early_stopping,
    verbose=True
)

# Сохраняем как лучшую модель
config_e4_full = {**model_config_e4, **train_config_e4}
save_experiment_result('E4', config_e4_full, history_e4, model=model_e4, is_best=True)
plot_history(history_e4, 'E4 EarlyStopping', FIGURES_DIR / 'curves_best.png')

best_exp_id = 'E4'
best_history = history_e4
best_model_config = model_config_e4.copy()
best_train_config = train_config_e4.copy()

# ============================================================================
# 10. ЧАСТЬ B (S09): ОПТИМИЗАЦИЯ (O1-O3)
# ============================================================================

print("\n" + "="*60)
print("ЧАСТЬ B: ОПТИМИЗАЦИЯ (O1-O3)")
print("="*60)

# Фиксированная архитектура для сравнения оптимизации (как в E4)
opt_model_config = {
    'hidden_dims': (256, 128),
    'dropout_p': 0.3,
    'use_batchnorm': True,
    'activation': 'ReLU',
}

criterion = nn.CrossEntropyLoss()

# ----------------------------------------------------------------------------
# O1: LR слишком большой
# ----------------------------------------------------------------------------
print("\n--- O1: LR слишком большой (1e-1) ---")
set_seed(SEED)

# Параметры ТОЛЬКО для обучения
train_config_o1 = {
    'optimizer': 'Adam',
    'lr': 1e-1,
    'weight_decay': 0,
    'momentum': 0,
}

model_o1 = MLP(INPUT_DIM, NUM_CLASSES, **opt_model_config).to(device)
optimizer_o1 = optim.Adam(model_o1.parameters(),
                           lr=train_config_o1['lr'],
                           weight_decay=train_config_o1['weight_decay'])
history_o1 = fit(model_o1, train_loader, val_loader, optimizer_o1, criterion, device,
                 epochs=8, verbose=True)

# Для отчёта объединяем конфиги
config_o1_full = {**opt_model_config, **train_config_o1}
save_experiment_result('O1', config_o1_full, history_o1)

# ----------------------------------------------------------------------------
# O2: LR слишком маленький
# ----------------------------------------------------------------------------
print("\n--- O2: LR слишком маленький (1e-5) ---")
set_seed(SEED)

train_config_o2 = {
    'optimizer': 'Adam',
    'lr': 1e-5,
    'weight_decay': 0,
    'momentum': 0,
}

model_o2 = MLP(INPUT_DIM, NUM_CLASSES, **opt_model_config).to(device)
optimizer_o2 = optim.Adam(model_o2.parameters(),
                           lr=train_config_o2['lr'],
                           weight_decay=train_config_o2['weight_decay'])
history_o2 = fit(model_o2, train_loader, val_loader, optimizer_o2, criterion, device,
                 epochs=8, verbose=True)

config_o2_full = {**opt_model_config, **train_config_o2}
save_experiment_result('O2', config_o2_full, history_o2)

# ----------------------------------------------------------------------------
# O3: SGD + momentum + weight decay
# ----------------------------------------------------------------------------
print("\n--- O3: SGD + momentum + weight_decay ---")
set_seed(SEED)

train_config_o3 = {
    'optimizer': 'SGD',
    'lr': 1e-2,
    'weight_decay': 1e-4,
    'momentum': 0.9,
}

model_o3 = MLP(INPUT_DIM, NUM_CLASSES, **opt_model_config).to(device)
optimizer_o3 = optim.SGD(
    model_o3.parameters(),
    lr=train_config_o3['lr'],
    momentum=train_config_o3['momentum'],
    weight_decay=train_config_o3['weight_decay']
)
history_o3 = fit(model_o3, train_loader, val_loader, optimizer_o3, criterion, device,
                 epochs=15, verbose=True)

config_o3_full = {**opt_model_config, **train_config_o3}
save_experiment_result('O3', config_o3_full, history_o3)

# ----------------------------------------------------------------------------
# График для O1 и O2 (плохие LR)
# ----------------------------------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loss
axes[0].plot(history_o1['train_loss'], label='O1 Train (LR=1e-1)', linestyle='--')
axes[0].plot(history_o1['val_loss'], label='O1 Val (LR=1e-1)', linestyle='--')
axes[0].plot(history_o2['train_loss'], label='O2 Train (LR=1e-5)', linestyle='-.')
axes[0].plot(history_o2['val_loss'], label='O2 Val (LR=1e-5)', linestyle='-.')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_title('Extreme LR Comparison - Loss')
axes[0].legend()
axes[0].grid(True)

# Accuracy
axes[1].plot(history_o1['val_acc'], label='O1 Val (LR=1e-1)', linestyle='--')
axes[1].plot(history_o2['val_acc'], label='O2 Val (LR=1e-5)', linestyle='-.')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('Extreme LR Comparison - Accuracy')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.savefig(FIGURES_DIR / 'curves_lr_extremes.png', dpi=150, bbox_inches='tight')
print(f"\nSaved: {FIGURES_DIR / 'curves_lr_extremes.png'}")
plt.show()

# ============================================================================
# 11. СОХРАНЕНИЕ runs.csv
# ============================================================================

print("\n" + "="*60)
print("СОХРАНЕНИЕ АРТЕФАКТОВ")
print("="*60)

runs_df = pd.DataFrame(all_runs)
runs_df.to_csv(ARTIFACTS_DIR / 'runs.csv', index=False)
print(f"Saved: {ARTIFACTS_DIR / 'runs.csv'}")
print("\nResults summary:")
print(runs_df[['experiment_id', 'best_val_accuracy', 'best_val_loss', 'epochs_trained']].to_string(index=False))

# ============================================================================
# 12. ФИНАЛЬНАЯ ОЦЕНКА ЛУЧШЕЙ МОДЕЛИ НА TEST
# ============================================================================

print("\n" + "="*60)
print("ФИНАЛЬНАЯ ОЦЕНКА НА TEST")
print("="*60)

# Загрузка лучшей модели
best_model = MLP(INPUT_DIM, NUM_CLASSES, **best_model_config).to(device)
best_model.load_state_dict(torch.load(ARTIFACTS_DIR / 'best_model.pt', map_location=device))

test_loss, test_acc = evaluate(best_model, test_loader, criterion, device)
print(f"\nBest Model ({best_exp_id}) on Test:")
print(f"  Test Loss: {test_loss:.4f}")
print(f"  Test Accuracy: {test_acc:.4f}")

# ============================================================================
# 13. СВОДКА
# ============================================================================

print("\n" + "="*60)
print("СВОДКА ЭКСПЕРИМЕНТОВ")
print("="*60)

print(f"\nДатасет: {DATA_CONFIG['dataset_name']}")
print(f"Лучший эксперимент: {best_exp_id}")
print(f"Лучшая val_accuracy: {best_val_acc:.4f}")
print(f"Тестовая accuracy: {test_acc:.4f}")
print(f"\nАртефакты сохранены в: {ARTIFACTS_DIR.absolute()}")
print(f"  - runs.csv")
print(f"  - best_model.pt")
print(f"  - best_config.json")
print(f"  - figures/curves_best.png")
print(f"  - figures/curves_lr_extremes.png")
```
